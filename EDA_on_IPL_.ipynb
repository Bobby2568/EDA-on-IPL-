##             Exploratory Data Analysis on 'Indian Premiere League'

```python
sdata=data1[['id','season','winner']]
sdata
```

```python
cdata=data2.merge(sdata,how='inner',left_on='match_id',right_on='id')
cdata.head().info()
```

```python
plt.figure(figsize = (18,10))
sns.countplot(x='city', data=data1)
plt.title('Most number of seasons hosted in city')
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize = (14,8))
sns.countplot(x='season', data=data1)
plt.title('Most number of matches played in a year')
plt.xticks(rotation='vertical')
plt.show()
```

--In the year most number of matches are played.



```python
toss= data1.toss_decision.value_counts()
labels = (np.array(toss.index))
sizes = (np.array((toss / toss.sum())*100))
colors = ['gold', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Toss decision percentage")
plt.show()
```

```python
num_of_wins = (data1.win_by_wickets>0).sum()
num_of_loss = (data1.win_by_wickets==0).sum()
labels = ["Wins", "Loss"]
total = float(num_of_wins + num_of_loss)
sizes = [(num_of_wins/total)*100, (num_of_loss/total)*100]
colors = ['gold', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Win percentage batting second")
plt.show()
```

```python
d1 = data1[['id','season','team1','team2','winner','toss_decision']]
second_batting_win = d1[(d1['toss_decision']== 'field')& (d1['team2']== d1['winner'])]
yearwins = second_batting_win['season'].value_counts().sort_index()
plt.bar(yearwins.index, yearwins)
plt.xlabel('Year')
plt.ylabel('Number of Wins')
plt.title('Teams Batting Second Wins Over the Years in IPL')
plt.show()
```

```python
d2_2019 = cdata[cdata['season'] == 2019]

# Calculate runs scored by hitting 6's and 4's for each batsman
d2_2019['runs_by_6s'] = d2_2019['batsman_runs'].eq(6).mul(d2_2019['batsman_runs'])
d2_2019['runs_by_4s'] = d2_2019['batsman_runs'].eq(4).mul(d2_2019['batsman_runs'])

# Group by batsman and sum the runs
total_runs_by_batsman = d2_2019.groupby('batsman')['runs_by_6s', 'runs_by_4s'].sum()

# Calculate total runs by adding 6's and 4's
total_runs_by_batsman['total_runs'] = total_runs_by_batsman['runs_by_6s'] + total_runs_by_batsman['runs_by_4s']

# Find the batsman with the highest total runs
highest_scorer = total_runs_by_batsman['total_runs'].idxmax()
print(f'In 2019 the batman who scored highest no of runs hitting 6 and 4 : {highest_scorer}')
```

```python
temp_df = cdata.groupby('bowler')['ball'].agg('count').reset_index().sort_values(by='ball', ascending=False).reset_index(drop=True)
temp_df = temp_df.iloc[:10,:]

labels = np.array(temp_df['bowler'])
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(10,8))
rects = ax.bar(ind, np.array(temp_df['ball']), width=width, color='cyan')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top Bowlers - Number of balls bowled in IPL")
ax.set_xlabel('Bowler Names')
plt.show()
```

```python
# We will print the top 10 leading run scorer in IPL
batsman_score=cdata.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score=batsman_score.rename(columns={'sum':'batsman_runs'})
print("*** Top 10 Leading Run Scorer in IPL ***")
batsman_score.iloc[:10,:]
```

```python
six_data=cdata[cdata['batsman_runs']==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

```python
batsman_six=six_data.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
ax=batsman_six.iloc[:10,:].plot('batsman','six',kind='bar',color='green')
plt.title("Numbers of six hit by playes ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No of six",fontsize=15)
plt.show()
```

```python
# we will print numbers of fours hit by team
four_data=cdata[cdata['batsman_runs']==4]
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python
# we will plot graph on four hit by players
batsman_four=four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax=batsman_four.iloc[:10,:].plot('batsman','four',kind='bar',color='black')
plt.title("Numbers of fours hit by playes ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Player name",fontsize=15)
plt.ylabel("No of fours",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='dismissal_kind', data=cdata)
plt.xticks(rotation='vertical')
plt.show()
```
